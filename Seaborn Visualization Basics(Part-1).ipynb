# Complete attribute documentation:

```python
# age: age in years
# sex: sex (1 = male; 0 = female)
# cp: chest pain type
                    #-- Value 1: typical angina
                    #-- Value 2: atypical angina
                    #-- Value 3: non-anginal pain
                    #-- Value 4: asymptomatic
# trestbps: resting blood pressure (in mm Hg on admission to the hospital)
# chol: serum cholestoral in mg/dl
# fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
# restecg: resting electrocardiographic results
                    #-- Value 0: normal
                    #-- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
                    #-- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
# thalach: maximum heart rate achieved
# exang: exercise induced angina (1 = yes; 0 = no)
# oldpeak = ST depression induced by exercise relative to rest
# slope: the slope of the peak exercise ST segment
                    #-- Value 1: upsloping
                    #-- Value 2: flat
                    #-- Value 3: downsloping
# ca: number of major vessels (0-3) colored by flourosopy
# thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
```

# Scatter plot and line plot In Seaborn:

```python
#Problem statement:
# I need to know which perticular age group is having cholestoral and blood sugar w.r.t male and female 
# One way:
sns.relplot(data=data,x="age",y="chol",col='gender',kind="scatter",hue='fps')
```

```python
#Another way
sns.relplot(data=data,x="age",y="chol",col='gender',kind="line",hue="fps")
```

```python
# In these plots i understand that, for male/female which age range of Patients having blood sugar, cholestoral
# 1: in case of female,maximum 50 to 65 aged people have higher Cholesterol levels, blood suger as well and in case of
# female when age is increasing the chol level is decresing.
# 2: in case of male.maxmimum 40 to 70 aged people have higher Cholesterol levels, blood suger as well and in case of male 
# when age is increasing chol level is increasing.
```

```python
sns.relplot(data=data,x="age",y="trestbps",kind="scatter",col="gender",hue='fps')
```

```python
sns.relplot(data=data,x="age",y="trestbps",kind="line",hue="fps",col="gender")
```

```python
# as we know, having 120/80 blood pressure is normal.above 120 is Risk.
# and in case of blood sugar till 180 is normal morethan is risk.
#in this plot,i got to know that for male/female which age range of Patients having blood sugar and blood pressure as well..
# 1. in case of female: maximum 40 to 65 aged people have blood sugar and blood pressure as well..
# 2. in case of male: maximum 40 to 70 aged people have blood sugar and blood pressure as well..
# 3. in this data, there are some males/females who are not having blood sugar and blood pressure
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(15,10))
sns.scatterplot(data=data,x="age",y="thalach",hue="gender",style="thal")
plt.show()
```

# Catplot

```python
# Categorical scatterplots:
            # stripplot() (with kind="strip"; the default)
            # swarmplot() (with kind="swarm")
# Categorical distribution plots:
            #  boxplot() (with kind="box")
            #violinplot() (with kind="violin")
            #boxenplot() (with kind="boxen")
#Categorical estimate plots:
            #pointplot() (with kind="point")
            #barplot() (with kind="bar")
            #countplot() (with kind="count")
```

```python
# stripplot
plt.style.use('dark_background')
sns.catplot(data=data,x="cp",y="trestbps",color="red",hue="gender",kind="strip")
plt.show()
```

```python
# swarmplot
plt.figure(figsize=(15,10))
sns.catplot(data=data,x="fps",y="trestbps",color="red",hue="gender",kind="swarm")
plt.show()
```

```python
#violinplot
sns.catplot(data=data,x='gender',y='thalach',kind='violin',color='red')
```

```python
sns.catplot(data=data,x='fps',y='trestbps',hue='thal',kind='violin',color='red',inner="stick")
plt.show()
```

```python
# boxplot
plt.style.use("tableau-colorblind10")
sns.catplot(data=data,x='thal',y='trestbps',hue='fps',col='gender',kind='box',color='red')
plt.show()
```

```python
#boxenplot
sns.boxenplot(data=data,x='thal',y='trestbps',hue='gender',color='red')
plt.show()
```

```python
# barplot
import numpy as np
sns.barplot(data=data,x="thal",y="oldpeak",estimator=np.mean)
```

```python
# jointplot
sns.jointplot(data=data,x='trestbps',y='chol',hue='gender')
plt.show()
```

```python
sns.jointplot(data=data,x='trestbps',y='thalach',hue='gender')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(15,10))
plt.style.use('seaborn-bright')
sns.heatmap(data=data.corr())
```

```python
# pointplot
plt.style.use('seaborn-bright')
sns.pointplot(data=data,x="gender",y='thalach',hue='slope')
plt.title("heartbit achived based on gender",fontdict={"size":20,"color":"red"},pad=20)
plt.xlabel("Gender",fontdict={"size":20,"color":"red"},labelpad=20)
plt.ylabel("thalach",fontdict={"size":20,"color":"red"},labelpad=20)
plt.show()
```

```python
# pairplot
plt.style.use('fivethirtyeight')
sns.pairplot(data,hue="gender")
```
